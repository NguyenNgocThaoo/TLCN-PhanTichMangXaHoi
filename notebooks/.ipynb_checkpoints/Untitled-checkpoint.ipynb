```python
from qdrant_client import QdrantClient
from qdrant_client.models import VectorParams, Distance

client = QdrantClient("qdrant", port=6333)  # dùng service name qdrant

# Tạo collection test
from qdrant_client.models import VectorParams, Distance

if not client.collection_exists("test_vectors"):
    client.create_collection(
        collection_name="test_vectors",
        vectors_config=VectorParams(size=4, distance=Distance.EUCLID),
    )

# Insert vector
client.upsert(
    collection_name="test_vectors",
    points=[{"id": 5, "vector": [0.2, 5, 0.3, 2], "payload": {"text": "hello qdrant"}}]
)



```

```python
# Query vector gần nhất
from qdrant_client.models import Filter, FieldCondition, MatchValue

res = client.query_points(
    collection_name="test_vectors",
    query=[0.1, 0.1, 0.3, 0.5],
    limit=2,
    # optional: filter
    query_filter=Filter(
        must=[FieldCondition(key="text", match=MatchValue(value="hello qdrant"))]
    )
)
print(res)
```

```python
from qdrant_client import QdrantClient

client = QdrantClient("qdrant", port=6333)

scroll_filter = None  # có thể thêm điều kiện filter nếu muốn
limit = 10
offset = None

while True:
    points, next_page = client.scroll(
        collection_name="test_vectors",
        scroll_filter=scroll_filter,
        limit=limit,
        offset=offset,
        with_vectors=True,      # <-- bật cái này để lấy luôn vector
        with_payload=True
    )

    for p in points:
        print(p.id, p.vector, p.payload)

    if next_page is None:
        break
    offset = next_page
```
